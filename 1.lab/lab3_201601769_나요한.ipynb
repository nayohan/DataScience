```python
import matplotlib.pyplot as plt
%matplotlib inline

from collections import Counter
import matplotlib

matplotlib.rc('font', family='Malgun Gothic')
##matplotlib.use('nbagg')
```

```python
def make_simple_line_chart():
    years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
    gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

    plt.plot(years, gdp, color='green', marker='v', linestyle='solid')
    plt.title("명목 GDP")
    plt.ylabel("Bilions of $")
    plt.show()
```

```python
make_simple_line_chart()
```

```python
def make_simple_bar_chart():
    movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
    num_oscars = [5, 11, 3, 8, 10]
    xs = [i+1 for i, _ in enumerate(movies)]
    plt.bar(xs,num_oscars)

    plt.title("My Favorite Movies")
    plt.ylabel("# of Academy Awards")
    plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies) 
    plt.show()
```

```python
make_simple_bar_chart()
```

```python
def make_histogram():
    grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
    decile = lambda grade:grade
    histogram = Counter(decile(grade) for grade in grades)  #grade에 grades에서 하나씩 뽑아서 넣어라.
    #print(histogram)
    #print(decile(grade))
    
    plt.bar([x for x in histogram.keys()], histogram.values(), 10)
    plt.axis([-5, 105, 0, 5])
    plt.xticks([10 * i for i in range(11)])
    plt.title("Distribution of Exam 1 Grades")
    plt.xlabel("Decile")
    plt.ylabel("# of Students")
    plt.show()
```

```python
make_histogram()
```

```python
def make_several_line_charts():
    variance     = [1,2,4,8,16,32,64,128,256]
    bias_squared = [256,128,64,32,16,8,4,2,1]
    total_error = [x+y for x,y in zip(variance, bias_squared)]
    
    xs = range(len(variance))
    plt.plot(xs, variance, 'g-', label='variance')
    plt.plot(xs, bias_squared, 'r-', label='bias*2')
    plt.plot(xs, total_error, 'b:', label='total error')
    
    plt.legend(loc=9)
    plt.xlabel("model complexity")
    plt.title("The Bias-Variance TradeOff")
    plt.show()
```

```python
make_several_line_charts()
```

```python
def make_scatter_plot():
    friends =  [70, 65, 72, 63, 71, 64, 60, 64, 67]
    minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
    labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
    
    plt.scatter(friends, minutes)
    for label, friend_count, minute_count in zip (labels, friends, minutes):
        plt.annotate(label, xy=(friend_count, minute_count), xytext=(5,-5), textcoords='offset points')
    
    plt.title("Daily Minutes vs. Number of Friends")
    plt.xlabel("# of friends")
    plt.ylabel("daily minutes spent on the site")
    plt.show()
```

```python
make_scatter_plot()
```

```python
import csv
import numpy as np
f = open('data.csv', 'r', encoding='utf-8')
rdr = csv.reader(f)
x = []
y = []
for line in rdr:
    x.append(line[0])
    y.append(line[1])
f.close()
npx = np.asarray(x)
npy = np.asarray(y)
npx = npx.astype(np.float)
npy = npy.astype(np.float)

print(max(npx), min(npx))
print(max(npy), min(npy))
```

```python
def make_visualize_scatter_data():
    plt.scatter(npx,npy)
    plt.title("2D Points Plotting")
    plt.axis([20, 80, 20, 120])
    
    plt.xticks([(i*10)+20 for i in range(6)])
    plt.yticks([(j*10)+20 for j in range(8)])
    plt.show()
```

```python
def make_visualize_bar_data():
    plt.bar(npx,npy, color="r")
    plt.title("2D Points Plotting")
    plt.axis([20, 80, 20, 120])
    
    plt.xticks([(i*10)+20 for i in range(6)])
    plt.yticks([(j*10)+20 for j in range(8)])
    plt.show()
```

```python
def make_visualize_plot_data():
    plt.plot(npx,npy, 'g', marker='o')
    plt.title("2D Points Plotting")
    plt.axis([20, 80, 20, 120])
    
    plt.xticks([(i*10)+20 for i in range(6)])
    plt.yticks([(j*10)+20 for j in range(8)])
    plt.show()
```

```python
make_visualize_scatter_data()
make_visualize_bar_data()
make_visualize_plot_data()
```

```python
def make_interpolation():
    plt.stem(npx, npy, 'r')
    plt.axis([20, 80, 20, 120])
    plt.xticks([(i*10)+20 for i in range(6)])
    plt.yticks([(j*10)+20 for j in range(8)])
    plt.show()
```

```python
def make_scatter_color():
    C = np.random.randint(0, 5, 100)
    print(C)
    plt.scatter(npx,npy, c=C, edgecolor='none', alpha=0.6, cmap=plt.cm.get_cmap('gist_rainbow', 6))
    
    plt.axis([20, 80, 20, 120])
    plt.xticks([(i*10)+20 for i in range(6)])
    plt.yticks([(j*10)+20 for j in range(8)])
    plt.show()
```
