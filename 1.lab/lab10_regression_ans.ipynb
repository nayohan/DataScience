```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

#기울기 a와 y 절편 b
wb=[3,76]

# x,y의 데이터 값
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

# y=ax + b에 a,b 값 대입하여 결과를 출력하는 함수
def predict(x):
    return wb[0]*x + wb[1]

# RMSE 함수
def rmse(a, p):
    return np.sqrt(((a - p) ** 2).mean())

# RMSE 함수를 각 y값에 대입하여 최종 값을 구하는 함수
def rmse_val(predict_result,y):
    return rmse(np.array(predict_result), np.array(y))

def step_gradient(w_current, b_current, points, learning_rate):
    #gradient descent
    w_gradient = 0
    b_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        w_gradient += -(2/N) * x * (y - (w_current * x + b_current)) # MSE 식의 w에 대한 편미분값 평균
        b_gradient += -(2/N) * (y - (w_current * x + b_current))     # MSE 식의 b에 대한 편미분값 평균
    new_w = w_current - (learning_rate * w_gradient) # w update
    new_b = b_current - (learning_rate * b_gradient) # b update
    return [new_w, new_b]
```

```python
# Define our Hyperparameters``
learning_rate = 0.03 #how fast the data converge

# y=ax+b (Slope formule)
# ab[0] initial slope guess
# ab[1] initial y-intercept guess

# 출력으로 확인
print("기울기 a =", wb[0])
print("y 절편 b =", wb[1])
    
yhat = wb[0] * np.array(x) + wb[1]
plt.plot(x, yhat, ':', color='green', label='guess')
    
for step in range(1000):
    print("\nstep =", step)
    # 예측값이 들어갈 빈 리스트
    predict_result = []
    # 모든 x값을 한 번씩 대입하여 predict_result 리스트완성.
    for i in range(len(x)):
        predict_result.append(predict(x[i]))
        print("공부시간=%.f, 실제점수=%.f, 예측점수=%.f" % (x[i], y[i], predict(x[i])))

    wb[0], wb[1] = step_gradient(wb[0], wb[1], np.array(data), learning_rate)
            
    # 최종 RMSE 출력
    print("rmse 최종값: " + str(rmse_val(y, predict_result)))
    
    # 출력으로 확인
    print("기울기 a =", wb[0])
    print("y 절편 b =", wb[1])
    
yhat = wb[0] * np.array(x) + wb[1]
plt.plot(x, yhat, '-', color='red', label='regress')
plt.legend()

plt.scatter(x, y)
plt.show()
```
