# 2-2번 문제

```python
import numpy as np

onehots = {}
onehots['cat'] = [1,0,0,0]
onehots['the'] = [0,1,0,0]
onehots['dog'] = [0,0,1,0]
onehots['sat'] = [0,0,0,1]
sentence = ['the','cat','sat']

def calc():
    list = np.array([0, 0, 0, 0])
    for i in range(len(sentence)):
        list += onehots[sentence[i]]
    return list

x = calc()
print("Sentence Encoding: " + str(x))
```
